# Acción 1: Aplicación modelos lineales.

```python
# Paso 1 Cargamos la base de datos

import pandas as pd
import os
import numpy as np

data=pd.read_csv("Anexo1-zomato.csv",engine='python',encoding="latin-1")

```

```python
# filtro

df1= data[(data["City"] == "New Delhi") & (data["Votes"] > 0)]

# Dividimos el conjunto de datos

n=len(df1)
prueba=int(n * 0.20) # 20% de datos para probar
entrenamiento=n-prueba # 80% de datos para entrenar el modelo 


train_data = df1[:entrenamiento].copy()  # datos de entrenamiento
prueba_data = df1[entrenamiento:].copy()  # datos de prueba


# convertimos la variable independiente categorica a numerica
train_data['Has Online delivery'] = train_data['Has Online delivery'].map({'Yes': 1, 'No': 0})
prueba_data['Has Online delivery'] = prueba_data['Has Online delivery'].map({'Yes': 1, 'No': 0})


# Definimos las variables predictoras para entrenar el modelo
x_train = train_data[['Average Cost for two', 'Votes', 'Has Online delivery']]
y_train = train_data['Aggregate rating']

# Definimos las variables predictoras para probar el modelo
x_test = prueba_data[['Average Cost for two', 'Votes', 'Has Online delivery']]
y_test = prueba_data['Aggregate rating']

```

```python
# cargar paquetes y aplicar el modelo lineal 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

mod = LinearRegression()
mod.fit(x_train, y_train)

print("Coeficientes:", mod.coef_)
print("Intercepto:", mod.intercept_)
```

```python
# Hacer predicciones utilizando el conjunto de prueba
y_pred = mod.predict(x_test)
# Calcular el error cuadrático medio (MSE) y el coeficiente de determinación (R²)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:",mse)
print("R^2:",r2)

```

# Acción 2: Aplicación modelo logístico.

```python
# creacion de la nueva variable

df1 = data[(data["City"] == "New Delhi") & (data["Votes"] > 0)].copy()

# Crear la columna 'nivel_calificacion' sin la advertencia
df1['nivel_calificacion'] = df1['Aggregate rating'].apply(lambda x: 'alto' if x > 3.7 else 'bajo')

```

```python
X = df1[['Average Cost for two', 'Votes', 'Has Online delivery']]
Y = df1['nivel_calificacion']

# Verificar el balance de la variable objetivo
print("Conteo de clases en 'nivel_calificacion':")
print(Y.value_counts())
```

```python
conteo_clases = Y.value_counts()
print("Conteo de clases en 'nivel_calificacion':")
print(conteo_clases)

# Si hay desbalance, hacer el balanceo con `iloc`
if conteo_clases['bajo'] != conteo_clases['alto']:  
    # Obtener el tamaño de la clase minoritaria
    menor_cantidad = conteo_clases.min()
    
    # Obtener índices aleatorios de ambas clases
    indices_alto = Y[Y == 'alto'].sample(n=menor_cantidad, random_state=42).index
    indices_bajo = Y[Y == 'bajo'].sample(n=menor_cantidad, random_state=42).index
    
    # Combinar los índices para obtener un conjunto balanceado
    indices_balanceados = np.concatenate([indices_alto, indices_bajo])
    
    # Seleccionar las filas balanceadas en X e y usando iloc
    X_balanced = X.loc[indices_balanceados]
    y_balanced = Y.loc[indices_balanceados]
else:
    X_balanced = X
    y_balanced = Y

# Verificar el balanceo final
print("Conteo de clases después del balanceo:")
print(y_balanced.value_counts())
```

```python
n = len(df1)
prueba_size = int(n * 0.20)  # 20% de datos para prueba
entrenamiento_size = n - prueba_size  # 80% de datos para entrenamiento

# Crear conjuntos de datos
datos_entrenamiento = df1[:entrenamiento_size].copy()  # datos de entrenamiento
datos_prueba = df1[entrenamiento_size:].copy()  # datos de prueba

# Convertir 'Has Online delivery' a valores numéricos
datos_entrenamiento['Has Online delivery'] = datos_entrenamiento['Has Online delivery'].map({'Yes': 1, 'No': 0})
datos_prueba['Has Online delivery'] = datos_prueba['Has Online delivery'].map({'Yes': 1, 'No': 0})


# Separar las variables independientes (X) y la variable objetivo (y)
X_train = datos_entrenamiento[['Average Cost for two', 'Votes', 'Has Online delivery']]
Y_train = datos_entrenamiento['nivel_calificacion']
X_test = datos_prueba[['Average Cost for two', 'Votes', 'Has Online delivery']]
Y_test = datos_prueba['nivel_calificacion']
```

```python
# ajuste del modelo logistico

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

mod2_logistico= LogisticRegression()
mod2_logistico.fit(X_train, Y_train)

# Predecir los valores en el conjunto de prueba
Y_pred = mod2_logistico.predict(X_test)
```

```python
# coeficientes
coeficientes = mod2_logistico.coef_
intercepto = mod2_logistico.intercept_

# Imprimir los coeficientes y el intercepto
print("Coeficientes del modelo:", coeficientes)
print("Intercepto del modelo:", intercepto)
```

```python
# Evaluación del modelo
precision = accuracy_score(Y_test, Y_pred)
# Imprimir los resultados
print("Precisión del modelo:", precision)
```

# Acción 3: Aplicación K-NN.

```python
# Teniendo en cuenta las variables definidas en el ejercisio anterior  entonces

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


```

Dado que el modelo es basado en aprendisaje automatico, La estrategia a utilizar para seleccionar el mejor modelo se basa en en la búsqueda por cuadricula.Esta estrategia tiene como objetivo considerar diferentes valores del parametro  vecino o n_neighbors (k) el cual se refiere al número de puntos de datos más cercanos que el algoritmo considera al hacer una predicción. Esto se hace con el fin de  encontrar la mejor combinación de estos en términos de su precision. Para escoger los valores del rango me base en el referente  Applied Predictive Modeling y Machine Learning de Jason Brownlee el cual sugiere  un equilibrio entre el ajuste insuficiente (cuando k es demasiado bajo) y el sobreajuste (cuando k es demasiado alto).

```python
modelo_knn = KNeighborsClassifier()
cuadricula= {'n_neighbors': range(1, 20)}  # Rango de valores de k 

# Configurar y ejecutar la búsqueda de hiperparámetros
grid_search = GridSearchCV(modelo_knn, cuadricula, cv=5)  # cv=5 realiza validación cruzada de 5 pliegues
grid_search.fit(X_train, Y_train)

# Mejor valor de k
mejor_k = grid_search.best_params_['n_neighbors']
print(f"El mejor valor de k es {mejor_k} con una precisión de {grid_search.best_score_:.4f}")
```

```python
knn_model_final = KNeighborsClassifier(n_neighbors=18)  

# Ajustar el modelo
knn_model_final.fit(X_train, Y_train)

# Realizar predicciones en el conjunto de prueba

y_predictor = knn_model_final.predict(X_test)

# Calcular la precisión y mostrar la matriz de confusión
accuracy = accuracy_score(Y_test, y_predictor)
conf_matrix = confusion_matrix(Y_test, y_predictor)

print("Precisión del modelo KNN:", accuracy)
print("Matriz de confusión:\n", conf_matrix)
```

# Acción 4: Aplicación Árboles de decisión (CART).



```python
# Teniendo en cuenta las variables definidas en el ejercisio anterior 
from sklearn.tree import DecisionTreeClassifier

modelo_arbol = DecisionTreeClassifier(random_state=42)
modelo_arbol.fit(X_train, Y_train)

# Hacer predicciones
predicciones = modelo_arbol.predict(X_test)

# Calcular la precisión
precision_arbol = accuracy_score(Y_test, predicciones)
print("Precisión del modelo de árbol de decisión:", precision_arbol)
```

```python
from sklearn.tree import export_graphviz
import graphviz

# Exportar el árbol
dot_data = export_graphviz(modelo_arbol, out_file=None, 
                           feature_names=['Average Cost for two', 'Votes', 'Has Online delivery'],  
                           class_names=['bajo', 'alto'],  
                           filled=True, rounded=True,  
                           special_characters=True)  
# Visualizar el árbol
grafico = graphviz.Source(dot_data)  
grafico.render("arbol_decision") 
grafico.view()  
```
